```python
#merging multi + ground truth + hyperspectral data
GS1_all_merged = pd.merge(merged_gt_multi, spectrum , how='outer')
```

```python
# Change the column 'REP' with 'Plant Number' since the replications are in the plot ids

GS1_all_merged.rename(columns={'REP':'Plant Number'}, inplace=True)
```

```python
# Isolating the replications from PLOT IDs
rep = []
for i in GS1_all_merged['PLOT ID']:
    str(i)
    if i[0] == '1' or i[0] == '2' :
        rep.append(i[0])
    else:
        rep.append(i[1])
```

```python
# inserting the real replications as a column
GS1_all_merged.insert(2, 'Rep', rep)
```

```python
#Checking the replication numbers
GS1_all_merged["Rep"].value_counts(dropna=False)
```

```python
# Isolating N treatments from PLOT ID
trt = []
for i in GS1_all_merged['PLOT ID']:
    str(i)
    if i[0] == '6':
        trt.append('HN')
    if i[0] == '5':
        trt.append('MN')
    if i[0] == '4':
        trt.append('LN')
    if i[0] == '1' or i[0] == '2':
        trt.append(np.NaN)
    
```

```python
# inserting the N treatments as a column
GS1_all_merged.insert(5, 'Trt', trt)
```

```python
# Checking the treatments
GS1_all_merged['Trt'].value_counts(dropna=False)
```

```python
# Export the final version of GS1 with all measurements
GS1_all_merged.to_csv('GS1_all_merged.csv')
```
