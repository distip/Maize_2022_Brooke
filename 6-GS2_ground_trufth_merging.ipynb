```python
import pandas as pd
import numpy as np
```

```python
data =pd.read_csv('GS2_processed_spectrum.csv')
gt = pd.read_csv('2022-Brooke-Maize_data-final.csv')
dw = pd.read_csv('Brooke_GS2_dryweights.csv')
```

```python
data.drop(data.filter(regex="Unname"),axis=1, inplace=True)
```

```python
data
```

```python
gt.rename(columns={'Plot_ID' : 'PLOT ID', 'Fresh_Weight' : 'Fresh_weight', 'Leaf_Area': 'Leaf_area', 'SPAD_1': 'CHL1', 
                  'SPAD_2 ' : 'CHL2', 'SPAD_3' : 'CHL3'},inplace=True)
```

```python
gt= gt.loc[gt['Growth_Stage'] == 2, :]
```

```python
merged = pd.merge(gt, data, how='outer')
```

```python
merged_v2 =pd.merge(dw, merged, how='outer')
```

```python
merged_v2
```

```python
SLA = merged_v2['Leaf_area'] / merged_v2['Dry weight']
```

```python
merged_v2.insert(6, 'SLA', SLA)
```

```python
LWC = (merged_v2['Fresh_weight']- merged_v2['Dry weight'])/merged_v2['Fresh_weight']*100
```

```python
merged_v2.insert(6, 'LWC', LWC)
```

```python
CHL = merged_v2[['CHL1', 'CHL2', 'CHL3']].mean(axis=1)
```

```python
merged_v2.insert(6, 'CHL', CHL)
```

```python
merged_v2.to_csv('Brooke_gt_merged')
```
