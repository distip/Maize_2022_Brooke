```python
#Remove the unnecessary row 'Unnamed: 0'

data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Remove the Leaf_pos column since we already got the average of it
data.drop('Leaf_pos', axis=1 , inplace=True)
```

```python
# Taking the three SPAD measurement average and inserting as a new column
data.insert(8, 'SPADmeter', data[['SPAD 1', 'SPAD 2', 'SPAD 3']].mean(axis=1))
```

```python
# remove the spad 1 spad 2 and spad 3 since we got their mean
data.drop(['SPAD 1', 'SPAD 2', 'SPAD 3'], axis=1, inplace=True)
```

```python
# There is another SPAD in the columns coming from multispeq we will change its name as 'SPAD_multispeq'
data.columns[0:40]
```

```python
#Changing the name of SPAD
data.rename(columns={'SPAD' : 'SPAD_multispeq'}, inplace=True)
```
